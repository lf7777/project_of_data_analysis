```python
import pandas as pd 
import numpy as np
from matplotlib import pyplot as plt
```

```python
# 导入数据
data = pd.read_csv('../../data/movie_metadata.csv')
# 设置显示所有列
pd.set_option('display.max_columns', None)
# 查看数据大小
data.shape
```

```python
# 分析列名的含义
data.columns.values.reshape(-1,1)
```

## 总体级别的数据清理

```python
# 删除不需要的列(专用单元格)
data = data.drop('content_rating',axis=1)
```

```python
# 保留有效值在18个以上的数据
data = data.dropna(thresh=18)
```

```python
# 查看剩余的数据大小
data.shape
```

```python
# 检测电影脸书喜欢数是否存在特殊离群值
data[data['movie_facebook_likes']>200000]
data[data['movie_facebook_likes']<0]
# 结论,数据无离群值
```

```python
# 检查电影上映时间有无离群值
data[data['title_year']>2021]
data[data['title_year']<1920]
# 结论,数据无离群值
```

```python
# 检测电影预算有无离群值
data[data['budget']>10000000000]
data[data['budget']<0]
# 结论,数据无离群值
```

```python
# 修改数据本身列名,改成更好理解的名字
data = data.rename(columns={'title_year':'release_dage','movie_facebook_likes':'facebook_likes'})
```

```python
# 填充导演姓名中的缺失值为 unknown_director_name
data.fillna(value={'director_name':'unknown_director_name'},inplace=True)
```

```python
# 填充缺失主演名字的行数据,为unknown_actor_1_name
data.fillna(value={'actor_1_name':'unknown_actor_1_name'},inplace=True)
```

## # 分析目标1,列出使票房收入较高的一些主角

```python
# 从总数据中,提取出主演姓名和电影的票房收入信息
actor_1 = data[['actor_1_name','gross']]
```

```python
# 首先清理数据
actor_1.info()
```

```python
# 删除缺失值(其中有一个为缺失,则删除)
actor_1 = actor_1.dropna(how='any')
actor_1
```

```python
# 检测离群值
# 检查最大值
actor_1.max()
```

```python
# 检查最小值
actor_1.min()
# 得出结论,数据无离群值
```

```python
actor_1.index = np.arange(4159)
```

```python
actor_1
```

```python
# 分析需求,最好使用平均数,找到平均票房稳定且高的主演
# 用名字分组,然后按照平均数排序,最后取前50个人,得到能使得票房比较稳定的前50个人的姓名及平均票房
actor_1.groupby('actor_1_name').mean().sort_values(by='gross',ascending=False).head(20)
```

```python
# CCH Pounder 排名15,成绩还不错,但是它的电影票房只有一部很高,拉高了平均数,可见其他数据还有和他相同的情况,或许我们还需要回归分析
actor_1[actor_1['actor_1_name']=='CCH Pounder']
```

```python
# 第一名只主演一部电影,但我们想要出演数量更稳定的演员,需要设置主演在榜单中出现次数大于3次
actor_1[actor_1['actor_1_name']=='Rupert Everett']
```

```python
a = actor_1.groupby('actor_1_name').count()['gross']
```

```python
# 将主演的出现次数和总票房合并
res = pd.concat([actor_1.groupby('actor_1_name').sum(),a],axis=1)
```

```python
res.columns=['gross','time']
res
```

```python
res = res[res['time']>=4]
# 此时得到了出演次数大于4的主演人名单
res
```

```python
res['mean']=res['gross']/res['time']
```

```python
res.sort_values(by='mean',ascending=False).head(30)
# 此时得出的数据,是出演次数大于4次,而且票房稳定在名单中前30的候选人
```

## 探索性分析,演员出演次数越多(经验)是否与票房平均呈正相关
### 使用散点图

```python
plt.scatter(res['time'],res['mean'])
plt.show()
```
