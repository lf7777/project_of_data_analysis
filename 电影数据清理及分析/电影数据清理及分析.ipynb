```python
# 导入数据
data = pd.read_csv('/home/lf/lll/data/data/movie_metadata.csv')

# 如果行数太多,默认省略部分列,此处设置显示所有列
pd.set_option('display.max_columns', None)

# 查看数据维度
data.shape
```

## 总体级别的数据清理

```python
# 删除不需要的列(专用单元格)
data = data.drop('content_rating',axis=1)
```

```python
# 保留有效值在18个以上的数据
data = data.dropna(thresh=18)
```

```python
# 查看剩余的数据大小
data.shape
```

### 检测离群值情况,判断数据是否真实

```python
# 检测电影脸书喜欢数是否存在特殊离群值
data[data['movie_facebook_likes']>200000]
data[data['movie_facebook_likes']<0]
# 结论,数据无离群值
```

```python
# 检查电影上映时间有无离群值
data[data['title_year']>2021]
data[data['title_year']<1900]
# 结论,数据无离群值
```

```python
# 检测电影预算有无离群值
data[data['budget']>10000000000]
data[data['budget']<0]
# 结论,数据无离群值
```

```python
# 修改数据本身列名,改成更好理解的名字
data = data.rename(columns={'title_year':'release_dage','movie_facebook_likes':'facebook_likes'})
```

### 通过数据维度得知,导演与主演的名字均有确实值情况,此处对缺失值进行处理

```python
# 填充导演姓名中的缺失值为 unknown_director_name
data.fillna(value={'director_name':'unknown_director_name'},inplace=True)
```

```python
# 填充缺失主演名字的行数据,为unknown_actor_1_name
data.fillna(value={'actor_1_name':'unknown_actor_1_name'},inplace=True)
```

### 分析目标1,列出使票房收入较高的一些主角

```python
# 从总数据中,提取出主演姓名和电影的票房收入信息
actor_1 = data[['actor_1_name','gross']]
```

### 清理数据

```python
# 首先清理数据,查看数据情况
actor_1.info()
```

```python
# 删除缺失值(其中有一个为缺失,则删除)
actor_1 = actor_1.dropna(how='any')
actor_1
```

```python
# 检测离群值
# 检查最大值
actor_1.max()
```

```python
# 检查最小值
actor_1.min()
# 得出结论,数据无离群值
```

```python
# 分析需求,最好使用平均数,出演次数需要4次以上(需求是稳定,单次出演肯定是不行的),找到票房平均数较高的主演
# 用名字分组,然后按照平均数排序,最后取前30个人
```

```python
# 求出演员在数据中出演的次数
time = actor_1.groupby('actor_1_name').count()['gross']
time
```

```python
# 将主演的出演的次数与总票房合并
res = pd.concat([actor_1.groupby('actor_1_name').sum(),time],axis=1)
```

```python
# 从新设置字段名称
res.columns=['gross','time']
res
```

```python
# 筛选出出演次数大于4的主演人名单和总票房
res = res[res['time']>=4]
res
```

```python
res['mean']=res['gross']/res['time']
```

```python
# 得出最终数据结果
pd.set_option('display.float_format', lambda x: '%.3f' % x)
res = res.sort_values(by='mean',ascending=False).head(30)
res['次序']=np.arange(1,31)
res
```

```python
# 进行可视化处理
res['gross'] = res['gross'].astype('str')
res['mean'] = res['mean'].astype('str')
res['gross'] = res['gross'].str.split('.').str[0]+'元'
res['mean'] = res['mean'].str.split('.').str[0]+'元'
```

```python
res
```

### 上面数据即是最终的名单可作为剧组选人的名单
### 此30位演员带来的票房收益是稳定且较高的

## 探索性分析,演员出演次数越多(经验)是否与票房平均呈正相关

```python
plt.scatter(res['time'],res['mean'])
plt.show()
```
