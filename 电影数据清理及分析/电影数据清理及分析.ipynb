```python
# 总体级别的数据清理
```

```python
# 删除不需要的列,专用单元格
data = data.drop('content_rating',axis=1)
```

```python
# 保留有效值在18个以上的数据
data = data.dropna(thresh=18)
```

```python
data.shape
```

```python
# 检测电影脸书喜欢数是否存在特殊离群值
data[data['movie_facebook_likes']>200000]
data[data['movie_facebook_likes']<0]
```

```python
# 检查电影上映时间有无离群值
data[data['title_year']>2021]
data[data['title_year']<1930]
```

```python
# 检测电影预算有无离群值
data[data['budget']>10000000000]
data[data['budget']<0]
```

```python
data = data.rename(columns={'title_year':'release_dage','movie_facebook_likes':'facebook_likes'})
```

```python
data.fillna(value={'director_name':'unknown_director_name'},inplace=True)
```

```python
data.fillna(value={'actor_1_name':'unknown_acrot_1_name'},inplace=True)
```

```python
# ------------------------------------------------------------------------------
```

```python
# 分析目标1,列出使票房收入较高的一些主角
actor_1 = data[['actor_1_name','gross']]
```

```python
actor_1.info()
```

```python
# 删除缺失值(其中有一个为缺失,则删除)
actor_1 = actor_1.dropna(how='any')
actor_1
```

```python
# 检测离群值
# 检查最大值
actor_1.max()
```

```python
# 检查最小值
actor_1.min()
# 得出结论,数据无离群值
```

```python
actor_1.index = np.arange(4155)
```

```python
actor_1
```

```python
# 用名字分组,然后按照平均数排序,最后取前50个人,得到能使得票房比较稳定的前50个人的姓名及平均票房
actor_1.groupby('actor_1_name').mean().sort_values(by='gross',ascending=False).head(50)
```
